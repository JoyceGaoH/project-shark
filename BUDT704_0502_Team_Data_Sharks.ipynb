#### 2.2 import the second dataset:

###3. Data cleaning
#### 3.1 Find and deal with Null-values

```python
df_shark_tank_original.isnull().sum()
```

The null values of Investment Amount and Investment Equity for each shark represents the shark did not invest in a specific start-up. Therefore, we can fill the null value in these columns to zero.

Drop cloums Pitchers Average Age, Entrepreneur Names and Company Website.

#### 3.2 Convert time to datetime format

```python
df_shark_tank_original["Season Start"]=pd.to_datetime(df_shark_tank_original["Season Start"])
df_shark_tank_original["Season End"]=pd.to_datetime(df_shark_tank_original["Season End"])
df_shark_tank_original["Original Air Date"]=pd.to_datetime(df_shark_tank_original["Original Air Date"])
df_shark_tank_original.info()
```

### 4. Data Manipulation: Merge two dataset to get text data

```python
df_shark_tank_s = df_shark_tank_s[['Pitched_Business_Identifier','Pitched_Business_Desc']]
df_shark_tank_s['Pitched_Business_Identifier_m'] = df_shark_tank_s['Pitched_Business_Identifier'].str.lower()
df_shark_tank_s['Name'] = df_shark_tank_s['Pitched_Business_Identifier_m'].str.replace('\s','',regex=True)
df_shark_tank_s
```

Convert the company name in both dataset to lower case as the cloumn to join on.

```python
df_shark_tank_original['Name'] = df_shark_tank_original['Startup Name'].str.lower()
df_shark_tank_merged = df_shark_tank_original.merge(df_shark_tank_s,how='left',on='Name')
df_shark_tank_merged
```

Since the longer pithes discription in the second dataset only contains Business Description information from season 1-8, we choose to combine the short Business Description from dataset one together with the longer description to one column.  

```python
df_shark_tank_merged['Business Description']=df_shark_tank_merged['Business Description'].fillna('').map(str)+'-'+df_shark_tank_merged['Pitched_Business_Desc'].fillna('').map(str)
df_shark_tank_merged
```
